```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/banking_data.csv')
print(df)
print('Data overview')
print(df.head())
print(df.info())
```

```python
#1.	What is the distribution of age among the clients?
plt.figure(figsize = (10,6))
sns.histplot(df['age'], bins=20, kde = True, color='blue')
plt.title('Age Distribution of Clients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#  2.How does the job type vary among the clients?
plt.figure(figsize=(10, 6))
job_counts = df['job'].value_counts()
plt.pie(job_counts, labels=job_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Job Type Variation Among Clients')
plt.legend()
plt.xlabel('Job type')
plt.ylabel('Count')
plt.show()
```

```python
# 3:What is the marital status distribution of the clients?
plt.figure(figsize=(8, 5))
marital_count = df['marital'].value_counts()
marital_count.plot(kind='bar', color=['skyblue', 'lightgreen', 'coral'])
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

```

```python

# 4: What is the level of education among the clients?
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='education', order=df['education'].value_counts().index, palette='coolwarm')
plt.title('Education Levels Among Clients')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

```python
# 5:	What proportion of clients have credit in default?
default_counts = df['default'].value_counts()
default_counts.plot(kind='bar', color=['red', 'green'], figsize=(8, 5))
plt.title('Proportion of Clients with Credit in Default')
plt.xticks(rotation=0)
plt.xlabel('Default Status')
plt.ylabel('Count')
plt.show()
```

```python
# 6: 	What is the distribution of average yearly balance among the clients?
plt.figure(figsize=(10, 6))
sns.histplot(df['balance'], bins=30, kde=True, color='purple')
plt.title('Distribution of Average Yearly Balance')
plt.xlabel('Balance (Euros)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 7: How many clients have housing loans?
housing_counts = df['housing'].value_counts()
housing_counts.plot(kind='bar', color=['orange', 'blue'], figsize=(8, 5))
plt.title('Number of Clients with Housing Loans')
plt.xticks(rotation=0)
plt.xlabel('Housing Loan')
plt.ylabel('Count')
plt.show()
```

```python
# 8: How many clients have personal loans?
loan_counts = df['loan'].value_counts()
loan_counts.plot(kind='bar', color=['pink', 'green'], figsize=(8, 5))
plt.title('Number of Clients with Personal Loans')
plt.xticks(rotation=0)
plt.xlabel('Personal Loan')
plt.ylabel('Count')
plt.show()
```

```python
# 9: What are the communication types used for contacting clients during the campaign?
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='contact', order=df['contact'].value_counts().index, palette='Set2')
plt.title('Communication Types Used During Campaign')
plt.xlabel('Contact Type')
plt.ylabel('Count')
plt.show()
```

```python
# 10:	What is the distribution of the last contact day of the month?
plt.figure(figsize=(10, 6))
sns.histplot(df['day'], bins=31, kde=False, color='teal')
plt.title('Last Contact Day of the Month Distribution')
plt.xlabel('Day of the Month')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 11: How does the last contact month vary among the clients?
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='month', order=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'], palette='muted')
plt.title('Last Contact Month Variation')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
# 12: What is the distribution of the duration of the last contact?
plt.figure(figsize=(10, 6))
sns.histplot(df['duration'], bins=30, kde=True, color='brown')
plt.title('Distribution of Duration of Last Contact')
plt.xlabel('Duration (Seconds)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 13: How many contacts were performed during the campaign for each client?
plt.figure(figsize=(10, 6))
sns.histplot(df['campaign'], bins=20, kde=False, color='lime')
plt.title('Number of Contacts During Campaign')
plt.xlabel('Number of Contacts')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 14: What is the distribution of the number of days passed since the client was last contacted from a previous campaign?
plt.figure(figsize=(10, 6))
sns.histplot(df['pdays'], bins=30, kde=False, color='orange')
plt.title('Number of Days Passed Since Last Contact')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 15: How many contacts were performed before the current campaign for each client?
plt.figure(figsize=(10, 6))
sns.histplot(df['previous'], bins=20, kde=False, color='red')
plt.title('Number of Contacts Before Current Campaign')
plt.xlabel('Number of Contacts')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 16:	What were the outcomes of the previous marketing campaigns?
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='poutcome', order=df['poutcome'].value_counts().index, palette='Blues')
plt.title('Outcomes of Previous Marketing Campaigns')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
# 17:	What is the distribution of clients who subscribed to a term deposit vs. those who did not?
subscribed_counts = df['y'].value_counts()
subscribed_counts.plot(kind='bar', color=['gray', 'green'], figsize=(8, 5))
plt.title('Subscription to Term Deposit')
plt.xticks(rotation=0)
plt.xlabel('Subscribed (yes/no)')
plt.ylabel('Count')
plt.show()
```

```python
# 18:	Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?
correlations = df.corr()
print("\nCorrelation with Subscription (y):")
print(correlations['balance'])

plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
